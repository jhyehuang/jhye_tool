```python
from  lib.mongo_db  import db_conn
import pandas as pd
import numpy as np
from pyecharts import Bar,configure,Pie,Line
import time
configure(output_image='svg')
```

```python
def date_to_time(date):
    timeArray = time.strptime(date, "%Y%m%d")
    time_semp = time.mktime(timeArray)
    return time_semp   # 2013

def time_to_today(timeStamp):
    timeArray = time.localtime(int(timeStamp))
    timeArray =time.strftime("%Y-%m-%d",timeArray)
    return timeArray   # 2013
```

```python
data=db_conn.sid_info.find({"end_time":{'$gte':date_to_time('20181109'),'$lt':date_to_time('20181116')}}
                           ,{'_id':0,'tel':1,'status':1,'end_time':1,'crawler_channel':1,'login_status':1,'status_report':1})
data1=list(data)
```

```python
len(data1)
```

```python
data1_rpt_pd=pd.DataFrame(x for x in data1)
```

```python
data1_rpt_pd['crawler_channel']=data1_rpt_pd['crawler_channel'].fillna('yulore')
```

```python
data1_rpt_pd['end_time']=data1_rpt_pd.end_time.map(time_to_today)
data1_rpt_pd=data1_rpt_pd.convert_objects(convert_numeric=True)
```

```python
data1_rpt_pd.shape
```

```python
def pd_group_by(data,col):
    data=data.groupby(['end_time','crawler_channel','tel']).size().reset_index()
    ret=data.groupby(['end_time','crawler_channel'])['tel'].size().reset_index()
    return ret.rename(index=str, columns={'tel':col})
```

```python
total_nums=pd_group_by(data1_rpt_pd,'尝试授权量')
```

```python
data2_rpt_pd=data1_rpt_pd.loc[data1_rpt_pd['login_status']==0]
authen_nums=pd_group_by(data2_rpt_pd,'授权成功量')
```

```python
data2_rpt_pd=data2_rpt_pd.loc[data2_rpt_pd['status']==0]
crawl_nums=pd_group_by(data2_rpt_pd,'crawl_nums')
```

```python
final_nums=pd_group_by(data2_rpt_pd,'爬取成功量')
```

```python
data2_rpt_pd=data2_rpt_pd.loc[data2_rpt_pd['status_report']==0]
report_nums=pd_group_by(data2_rpt_pd,'报告生成量')
```

```python
import copy
crawl_tel_nums=copy.deepcopy(crawl_nums)
crawl_tel_nums=crawl_tel_nums.rename(index=str, columns={'crawl_nums':'爬取成功号码量'})
```

```python
all_nums=pd.merge(total_nums,authen_nums,how='left',on=['crawler_channel','end_time'])
#all_nums=pd.merge(all_nums,crawl_nums,how='left',on=['crawler_channel','end_time'])
all_nums=pd.merge(all_nums,final_nums,how='left',on=['crawler_channel','end_time'])
all_nums=pd.merge(all_nums,report_nums,how='left',on=['crawler_channel','end_time'])
all_nums=pd.merge(all_nums,crawl_tel_nums,how='left',on=['crawler_channel','end_time'])
```

```python
all_nums['授权成功率']=all_nums['授权成功量']/all_nums['尝试授权量']
all_nums['爬取成功率']=all_nums['爬取成功量']/all_nums['授权成功量']
```

```python
all_nums
```

```python
def pri_tu(data,col):
    bar= Line('day_anly','Theme',renderer ='svg')
    for values in data.crawler_channel.values.tolist():
        sub_data=data.loc[data.crawler_channel.values==values,:]
        str1=sub_data.end_time.values.tolist()
        #str1.sort()
        bar.add(values,str1,sub_data[col].values.tolist(),yaxis_min=round(min(sub_data[col].values.tolist()),2),is_more_utils = True,is_smooth=True)
    return bar
```

```python
pri_tu(all_nums,'授权成功率')
```

```python
pri_tu(all_nums,'爬取成功率')
```
